```python
df = pd.read_csv('telecom.csv', usecols = ['OnlineSecurity','OnlineBackup','Churn'])
```

```python
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(df['OnlineSecurity'])
```

```python
plt.hist(df['OnlineBackup'])
```

```python
plt.hist(df['Churn'])
```

```python
plt.figure(figsize=(14, 10))
sns.countplot(data=df, x='OnlineSecurity', hue='Churn')
plt.title('Churn based on OnlineSecurity')
plt.xlabel('OnlineSecurity')
plt.ylabel('Count')
plt.legend(title='Churn', labels=['Not Churned', 'Churned'])
plt.show()
```

```python
plt.figure(figsize=(14, 10))
sns.countplot(data=df, x='OnlineBackup', hue='Churn')
plt.title('Churn based on OnlineBackup')
plt.xlabel('OnlineBackup')
plt.ylabel('Count')
plt.legend(title='Churn', labels=['Not Churned', 'Churned'])
plt.show()
```
